```python
import pandas as pd
import numpy as np
```

```python
tnx = pd.read_csv(r"C:\My Data Analyst Journey\Datasets\lampay\transactions_data.csv")
```

```python
cards = pd.read_csv(r"C:\My Data Analyst Journey\Datasets\lampay\cards_data.csv")
```

```python
user = pd.read_csv(r"C:\My Data Analyst Journey\Datasets\lampay\users_data.csv")
```

```python
#preprocessing Transactions Data set
print('transactions info',tnx.info())
print('No.of duplicated rows',tnx.duplicated().sum())
print('Sum of null values',tnx.isnull().sum())
print('transactions data shape',tnx.shape)
tnx['merchant_state']=tnx['merchant_state'].fillna('ONLINE')#filling the null values
tnx['zip']=tnx['zip'].fillna(0)# filling the null values
tnx['errors']=tnx['errors'].fillna('No Error') # filling the null values
print('Sum of null values',tnx.isnull().sum())# after handlling null values make sure there is no null values
tnx.drop(columns=['errors'],inplace = True)# coloun dropp
print('transactions Data preprocessing done')
```

```python
#preprocessing Cards Data set
print('cards info',cards.info())
print('No.of duplicated rows',cards.duplicated().sum())
print('Sum of null values',cards.isnull().sum())
print('cards data shape',cards.shape)
cards=cards.drop(columns=['card_number','cvv',])
cards.rename(columns={'id':'card_id'},inplace =True)
print('Cards Data preprocessing done')
```

```python
#preprocessing User Data set
print('user info',user.info())
print('No.of duplicated rows',user.duplicated().sum())
print('Sum of null values',user.isnull().sum())
print('user data shape',user.shape)
user = user.drop(columns=['birth_year','birth_month','address'])
user.rename(columns={'id':'client_id'},inplace= True)
print('User Data preprocessing done')
```

```python
# Merging Cards Data with transaction data
print('transaction shape before merging',tnx.shape)
tnx_1 = tnx.merge(cards,on=['card_id','client_id'], how ='left')
print('transaction shape After merging',tnx_1.shape)
```

```python
# Merging User data with merged transactions and Cards data
print('transaction shape before merging',tnx_1.shape)
final_df = tnx_1.merge(user,on='client_id', how  ='left')
print('transaction shape before merging',final_df.shape)
''' Checking shape of the dataset after merging to ensure no duplication happened after merging data '''
```

 ## Stratified Sampling

```python
''' Sampling data as the raw dataset is too large for analysis, using Stratified Sampling for sampling the data 
'frac=0.1' to get 10% of the data and using 'random_state=42' to reproduce the same result when re executed '''
```

```python
print('Dataset shape before sampling',final_df.shape)
ftd=(final_df.groupby(['card_type','mcc'],group_keys=False).apply(lambda x: x.sample (frac=0.1,random_state=42)))
print('Dataset shape after sampling',ftd.shape)
```

```python
''' Re- Sampling the sample data it exceed the excel sheet limit'''
```

```python
ftd1 = (ftd.groupby(['card_type','mcc'],group_keys=False).apply(lambda x: x.sample(frac=0.7,random_state=42)))
print('Dateset shape after re-sampling',ftd1.shape)
```

```python
ftd1['Date'] = pd.to_datetime(ftd1['date']).dt.date
ftd1['time'] = pd.to_datetime(ftd1['date']).dt.time
ftd1.drop(columns=['date'],inplace=True)
```
